```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns





```

```python


x:list[str]=['EDA','Data Wrangling','Mineria datos','Limpieza Datos','SQL','Est. Inferencial','ML','DL','IA']
y:list[int]= [5,10,15,20,30,25,31,30,20]

data = {
    "Conocimientos previos":x,
    "Frecuencia": y
}

dataset = pd.DataFrame(data)
dataset




```

```python
sns.set_theme(style="whitegrid", palette="pastel")
plt.figure(figsize=(16,9))
sns.barplot(dataset, x="Conocimientos previos", y="Frecuencia")
plt.show()
```
